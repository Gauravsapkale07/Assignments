Checking and Treating oultiers

```python
plt.figure(figsize=(16,9))
plt.boxplot(data,labels = data.columns);
```

```python
def capping(data,columns):
    for i in columns:
        Q1 = data[i].quantile(0.25)
        Q3 = data[i].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - IQR * 1.5 
        upper_bound = Q3 + IQR * 1.5
        # directly changes the original data  
        data.loc[:, i] = data[i].clip(lower=lower_bound, upper=upper_bound) 
```

```python
capping(data,['ASTV','Width'])
```

```python
plt.figure(figsize=(16,9))
plt.boxplot(data,labels = data.columns);
```

IQR Calculation

```python
for col in data.columns:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    iqr = Q3 - Q1 
    print(f"Column: {col} has IQR: {iqr}")
```
